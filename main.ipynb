```python
# Step 1: Importing database

import pandas as pd

tabela = pd.read_csv("cancelamentos.csv")

# kaggle (site para bases de dados)
# Step 2: Visualizing database

tabela = tabela.drop(columns="CustomerID")
display(tabela)
```

```python
# Step 3: Fixing problems in the database
display(tabela.info())

# removing null values
tabela = tabela.dropna()

display(tabela.info())
```

```python
# Step 4: Initial analysis (how many clients have canceled)

# Amount
display(tabela["cancelou"].value_counts())

# Porcentage
display(tabela["cancelou"].value_counts(normalize=True))
# display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
# Passo 5: Detailed Analysis (reason why client canceled, how each column impacts customer cancellation)
import plotly.express as px

for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna, color="cancelou", text_auto=True)

    grafico.show()
```
